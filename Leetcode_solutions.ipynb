```python
# most occurance
def Majority(nums):
  candidate=None
  count=0
  for num in nums:
    if count==0:
      candidate=num
    # count+=1 if num==candidate else -1
    if num==candidate:
      count+=1
    else:
      count-=1
  return candidate

nums=[1,4,2,3,2,1,2,2,1]
Majority(nums)
```

```python
def Palindrome(num):
  orgi=num
  rev=0
  while num>0:
    rem=num%10
    rev=rev*10+rem
    num=num//10
  if rev==orgi:
    return True
  return False
print(Palindrome(1221))
```

```python
# Roman to Integer
def RomantoInt(str1):
  lookup={
      "I" : 1,
      "V" : 5,
      "X" : 10,
      "L" : 50,
      "C" : 100,
      "D" : 500,
      "M" : 1000
  }

  s=len(str1)
  out=0
  i=s-1
  while i>=0:
    if i<s-1 and lookup[str1[i]]<lookup[str1[i+1]]:
      out-=lookup[str1[i]]
    else:
      out+=lookup[str1[i]]
    i=i-1
  return out

print(RomantoInt("MCMXCIV"))

```

```python
def longest_common_prefix(strs):
    if not strs:
        return ""
    strs.sort()

    first = strs[0]
    last = strs[-1]

    prefix = ""
    for i in range(len(first)):
        if  first[i] == last[i]:
            prefix += first[i]
        else:
            break

    return prefix
strs = ["flower","flow","flight"]
print(longest_common_prefix(strs))
```

```python
def valid_paren(str1):
  l=[]
  for i in str1:
    if i=='(' or i=='{' or i=='[':
      l.append(i)
    else:
      if i ==')' and l[-1]=='(':
        l.pop()
      elif  i =='}' and l[-1]=='{':
        l.pop()
      elif  i ==']' and l[-1]=='[':
        l.pop()
      else:
        return False
  print(l)
  if len(l)==0:
    return True
  else:
    return False

str1="({})]"
print(valid_paren(str1))
```

```python
def valid_parent(strs):
  l=[]
  for i in strs:
    if i =="(" or i =="{" or i =="[":
      l.append(i)
    else:
      if i ==")":
        if l[-1]=="(":
          l.pop()
        else:
          return False
      if i =="}":
        if l[-1]=="{":
          l.pop()
        else:
          return False
      if i =="]":
        if l[-1]=="[":
          l.pop()
        else:
          return False

  if len(l)==0:
    return True
  else:
    return False

str1="({}))"
print(valid_parent(str1))
```

```python
# remove duplicates from a set of numbers
def duplicate(arr):
  j=1
  for i in range(1,len(arr)):
    if arr[i]!=arr[i-1]:
      arr[j]=arr[i]
      j+=1
  return j

nums = [0,0,1,1,1,2,2,3,3,4]
print(duplicate(nums))
```

```python
from queue import deque
def similar(s):
  s=list(s)
  q=deque(s)
  while q:
    if q[0]==q[-1]:
      q.pop()
      q.popleft()
    else:
      break
  return q
s="abcaba"
print(similar(s))
```

```python
def intersection(num1,num2):
  num3=[]
  for i in num1:
    for j in num2:
      if i in num3:
        break
      if i==j:
        num3.append(i)
        break
  return num3
num1=[4,9,5]
num2=[9,4,9,8,4]
print(intersection(num1,num2))
```

```python
def intersection2(num1,num2):
  num3=[]
  for i in num1:
    if i in num3:
      continue
    if i in num2:
      num3.append(i)
      continue
  return num3
num1=[4,9,5]
num2=[9,4,9,8,4]
print(intersection2(num1,num2))
```

```python
def searchInsert(nums,target):
  l,r=0,len(nums)-1
  while l<=r:
    mid=(l+r)//2
    if target==nums[mid]:
      return mid
    if target<nums[mid]:
      r=mid-1
    elif target>nums[mid]:
      l=mid+1
  return l

nums=[-1,3,5,6]
print(searchInsert(nums,0))
```
